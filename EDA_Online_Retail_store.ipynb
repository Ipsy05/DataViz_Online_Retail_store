#### Bar Chart: Visualizing Sales Distribution Among Product Categories

A bar chart provides a visual comparison of sales figures across distinct product categories. It employs numerical values on one axis and category descriptions on the other, offering a straightforward representation of the sales distribution within the specified product categories.


```python
# Data for the bar chart
product_category = np.array(['Furniture', 'Technology', 'Office Supplies'])
sales = np.array([4110451.90, 4744557.50, 3787492.52])

# Define colors for each product category
colors = ['brown', 'blue', 'orange']

# Create a bar chart
plt.bar(product_category, sales, color=colors)

# Display the bar chart
plt.show()
```

##### Bar Chart - Plot sales across each product category

1. Adding labels to Axes
2. Reducing the bar width
3. Giving Title to the chart
4. Modifying the ticks to show information in (million dollars)

```python
# Data for the bar chart
product_category = np.array(['Furniture', 'Technology', 'Office Supplies'])
sales = np.array([4110451.90, 4744557.50, 3787492.52])

# Define colors for each product category
colors = ['brown', 'blue', 'orange']

# Create a bar chart with adjusted parameters
plt.bar(product_category, sales, color=colors, width=0.5)  # Reducing bar width for better clarity

# Adding labels to Axes
plt.xlabel('Product Categories')
plt.ylabel('Sales (Million Dollars)')

# Giving Title to the chart
plt.title('Sales Across Product Categories')

# Modifying the ticks to show information in million dollars
plt.ticklabel_format(style='plain', axis='y')  # Disable scientific notation on the y-axis
plt.yticks(np.arange(0, max(sales) + 1000000, 1000000), ['${:.2f}M'.format(x / 1000000) for x in np.arange(0, max(sales) + 1000000, 1000000)])

# Display the bar chart
plt.show()
```

##### Subplots are shown in seperate graphs. Each chart can be of different types.

```python
# Creating subplots with a 2x3 grid
fig, ax = plt.subplots(ncols=3, nrows=2, sharex=True, sharey=True)

# Plotting sales data for Europe in the first subplot
europe, = ax[0][0].plot(years, sales_Europe)
europe.set_label("Europe")

# Plotting bar chart for USCA in the second subplot
usca = ax[0][1].bar(years, sales_USCA)
usca.set_label("USCA")

# Plotting scatter plot for Africa in the third subplot
africa = ax[0][2].scatter(years, sales_africa)
africa.set_label("Africa")

# Plotting sales data for Asia Pacific in the fourth subplot
asia, = ax[1][0].plot(years, sales_Asia_Pacific)
asia.set_label("Asia Pacific")
asia.set_dashes([2, 2, 10, 2])  # Setting custom dash pattern

# Plotting sales data for LATAM in the fifth subplot
latam, = ax[1][1].plot(years, sales_LATAM)
latam.set_label("LATAM")
latam.set_dashes([2, 5, 5, 2])  # Setting custom dash pattern

# Removing the sixth subplot (ax[1][2]) as it is not used

# Display the plot
plt.show()
```

```python
# Creating subplots with a 1x4 grid and sharing the y-axis
fig, ax = plt.subplots(ncols=4, sharey=True)

# Plotting sales data for Europe in the first subplot
europe, = ax[0].plot(years, sales_Europe)
europe.set_label("Europe")
europe.set_color("red")  # Setting line color
ax[0].set_title('Sales in Europe')  # Setting subplot title

# Plotting bar chart for USCA in the second subplot
usca = ax[1].bar(years, sales_USCA)
usca.set_label("USCA")
ax[1].set_title('Sales in USCA')  # Setting subplot title

# Plotting scatter plot for Africa in the third subplot
africa = ax[2].scatter(years, sales_africa)
africa.set_label("Africa")
ax[2].set_title('Sales in Africa')  # Setting subplot title

# Plotting bar charts for Asia Pacific and LATAM in the fourth subplot
asia = ax[3].bar(years, sales_Asia_Pacific, width=0.5, color='royalblue')
latam = ax[3].bar(years, sales_LATAM, width=0.5, color='seagreen')
ax[3].set_title('Sales in Asia Pacific and LATAM')  # Setting subplot title
ax[3].legend((asia[0], latam[0]), ('Asia Pacific', 'LATAM'))  # Adding legend

# Adjusting the size of the figure
fig.set_size_inches(20.5, 5.5, forward=True)

# Display the plot
plt.show()
```
